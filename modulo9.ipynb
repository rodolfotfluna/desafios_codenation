```python
import pandas as pd
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
#Armazenando as inscrições para serem utilizadas no arquivo de respostas.

df_resp = pd.DataFrame()

df_resp['NU_INSCRICAO'] = df_test['NU_INSCRICAO']
```

```python
df_test.columns
```

```python
df_train['IN_TREINEIRO'].value_counts()
```

```python
df_train.isnull().sum(axis = 0)
```

```python
df_train[df_train.isna().any(axis=1)]
```

```python
#Excluindo colunas do dataset de treino que não estão no datset de teste.

col = list(df_test.columns)
col.append('IN_TREINEIRO')

df_train = df_train[col]
```

```python
#Vamos excluir todas as colunas com dados faltantes e testar se temos bons resultados.

df_train = df_train.dropna(axis='columns')
```

```python
df_train.head()
```

```python
#Vamos selecionar apenas as colunas com correlação maiores que 0.3.
pearsoncorr = df_train.corr(method='pearson')
var = pearsoncorr.columns[(pearsoncorr.IN_TREINEIRO>0.30) | (pearsoncorr.IN_TREINEIRO<-0.30)] 
var = var.tolist()
var
```

```python
df_train = df_train[var]
df_train.head()
```

```python
df_test = df_test[df_train.columns[:-1]]
df_test.head()
```

```python
#Dividindo os dados.

y = df_train['IN_TREINEIRO']
X = df_train.drop(['IN_TREINEIRO'], axis=1)
```

```python
#Criando o modelo random forest.

from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(random_state=0).fit(X, y)
```

```python
pred = clf.predict(df_test)
```

```python
df_resp['IN_TREINEIRO'] = pred
```

```python
df_resp.head()
```

```python
df_resp.to_csv('answer.csv',index=False)
```

```python
#SCORE=96.10
```

```python
#Balanceando os dados para tentar melhorar a performance.

from imblearn.over_sampling import SMOTE

sm = SMOTE(sampling_strategy='auto', k_neighbors=1, random_state=100)

X_res, y_res = sm.fit_resample(X,y)
```

```python
#Criando modelo com os dados balanceados

clf_res = RandomForestClassifier(random_state=0).fit(X_res, y_res)
```

```python
pred_res = clf_res.predict(df_test)
```

```python
df_resp['IN_TREINEIRO'] = pred_res
df_resp.head()
```

```python
df_resp.to_csv('answer.csv',index=False)
```
